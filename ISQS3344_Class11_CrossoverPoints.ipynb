## Simple plot
In this section, we want to draw the software a cost and software a cost functions on the same plot. Starting from the default settings, we’ll enrich the figure step by step to make it nicer.

Matplotlib comes with a set of default settings that allow customizing all kinds of properties. You can control the defaults of almost every property in matplotlib: figure size and dpi, line width, color and style, axes, axis and grid properties, text and font properties and so on.



First step is to get the data for the software a cost and software a cost functions:

```python
# generate the software costs based on report_volume

a_cost = 200000 + (60 * report_volume)
b_cost = 300000 + (25 * report_volume)
c_cost = 400000 + (10 * report_volume)


# generate the software a costs and software a costs of report_volume values
# a_cost = np.cos(report_volume)
# b_cost = np.sin(report_volume)
```

`report_volume` is now a `NumPy` array with 41 values ranging from -π to +π (included).      
`a_cost` is an array of the software a costs of `report_volume` (41 values).    
`b_cost` is an array of the software b costs of `report_volume` (41 values).    
`c_cost` is an array of the software c costs of `report_volume` (41 values). 

####Plot the software a cost and software a cost of `report_volume` values    

Using the three arrays we have created, we can use `matplotlib`'s `plot` method to essentially create two overlaying plots:    
1. a plot of `report_volume` values on the **x-axis** and their software a costs on the **y-axis**    
2. a plot of `report_volume` values on the **x-axis** and their software b costs on the **y-axis**  
3. a plot of `report_volume` values on the **x-axis** and their software c costs on the **y-axis**  


```python
# use maplotlib plot to chart software a cost and software a cost for each report_volume value
plt.plot(report_volume, a_cost)  # the blue line in the chart 
plt.plot(report_volume, b_cost)  # the red line in the chart
plt.plot(report_volume, c_cost)  # the red line in the chart


# Show the plot as a result on the default screen
plt.show()
```

## Setting ticks
Current ticks are not ideal because they do not show the interesting values (+/-π,+/-π/2) for software a cost and software a cost. We’ll change them such that they show only these values.     

**More information available:**    
- [xticks() command](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html?highlight=xticks#matplotlib.pyplot.xticks)
- [yticks() command](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html?highlight=yticks#matplotlib.pyplot.yticks)
- [Tick container](https://matplotlib.org/stable/api/container_api.html?highlight=artists+container)
- [Tick locating and formatting](https://matplotlib.org/stable/api/ticker_api.html?highlight=ticker+api)

```python
# Create a figure of size 8x6 (WidthxHeight) inches, 80 dots per inch (DPI)
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot in a grid of 1x1 (RowsxColumns) are arg1, arg2
# arg3 is the subplot index number (position in the grid, top to right)
plt.subplot(1, 1, 1)

# Plot software a cost with a blue continuous line of width 2 (pixels) ### add label for legend!
plt.plot(report_volume, a_cost, color="blue", linewidth=2.0, linestyle="-", label="Software A cost")
# Plot software a cost with a red dashed line of width 2 (pixels) ### add label for legend!
plt.plot(report_volume, b_cost, color="red", linewidth=2.0, linestyle="--", label="Software B cost")
# Plot software a cost with a red dashed line of width 2 (pixels)
plt.plot(report_volume, c_cost, color="green", linewidth=2.0, linestyle=":", label="Software C cost")

# inserting a legend in the upper-left of the plot
plt.legend(loc='upper left')


# Set x-axis value limits
plt.xlim(0.0, 20000.0)

# Set x-axis ticks (placement of tick marks on x-axis)
plt.xticks(np.linspace(0, 20000, 9, endpoint=True))

# Set y-axis value limits
plt.ylim(0.0, 1500000)

# Set y-axis ticks (placement of tick marls on y-axis)
plt.yticks(np.linspace(0, 1500000, 7, endpoint=True))

# Save figure as a PNG graphic file using 72 dots per inch
# comment out the following line if no `/content/` folder available
plt.savefig("/content/class_11_crossover2.png", dpi=72)

# Show the plot as a result on the default screen
plt.show()
```

### Add crossover points to x axis

```python
# Create a figure of size 8x6 (WidthxHeight) inches, 80 dots per inch (DPI)
plt.figure(figsize=(8, 6), dpi=80)

# Create a new subplot in a grid of 1x1 (RowsxColumns) are arg1, arg2
# arg3 is the subplot index number (position in the grid, top to right)
plt.subplot(1, 1, 1)

# Plot software a cost with a blue continuous line of width 2 (pixels) ### add label for legend!
plt.plot(report_volume, a_cost, color="blue", linewidth=2.0, linestyle="-", label="Software A cost")
# Plot software a cost with a red dashed line of width 2 (pixels) ### add label for legend!
plt.plot(report_volume, b_cost, color="red", linewidth=2.0, linestyle="--", label="Software B cost")
# Plot software a cost with a red dashed line of width 2 (pixels)
plt.plot(report_volume, c_cost, color="green", linewidth=2.0, linestyle=":", label="Software C cost")

# inserting a legend in the upper-left of the plot
plt.legend(loc='upper left')

# adding annotation: dashed blue line for software a cost crossover
a_volume_cross = 2857   # ($60 * 2,857 report volume plus $200,000 = $371,420)
a_cost_cross = ((60 * a_volume_cross) + 200000)


# adding line: dashed blue line 
plt.plot([a_volume_cross, a_volume_cross],[a_volume_cross, a_cost_cross], color='blue', linewidth=2.5, linestyle="--")
plt.scatter([a_volume_cross, ],[a_cost_cross, ], 50, color='red')


# adding annotation: dashed red line for software b cost crossover
b_volume_cross = 6666   # ($25 * 6,666 report volume plus $300,000 = $466,650)
b_cost_cross = ((25 * b_volume_cross) + 300000)

# adding line: dashed red line 
plt.plot([b_volume_cross, b_volume_cross],[b_volume_cross, b_cost_cross], color='red', linewidth=2.5, linestyle="--")
plt.scatter([b_volume_cross, ],[b_cost_cross, ], 50, color='green')

# Set x-axis value limits
plt.xlim(0.0, 20000.0)

# Set x-axis ticks (placement of tick marks on x-axis)
plt.xticks(np.linspace(0, 20000, 9, endpoint=True))

# Set y-axis value limits
plt.ylim(0.0, 1500000)

# Set y-axis ticks (placement of tick marls on y-axis)
plt.yticks(np.linspace(0, 1500000, 7, endpoint=True))

# Save figure as a PNG graphic file using 72 dots per inch
# comment out the following line if no `/content/` folder available
plt.savefig("/content/class_11_crossover3.png", dpi=72)

# Show the plot as a result on the default screen
plt.show()
```
